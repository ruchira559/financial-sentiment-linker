```python
import pandas as pd
from newsapi import NewsApiClient
import yfinance as yf
import spacy
import re
import torch
```

```python
# 1. Initialize API (Replace 'YOUR_API_KEY' with your actual key)
newsapi = NewsApiClient(api_key='e5627542cb9e4af793e97a828ec134aa')

def get_company_news(query):
    # Fetch headlines about the company from the last 30 days
    articles = newsapi.get_everything(q=query,
                                      language='en',
                                      sort_by='relevancy',
                                      page_size=20)

    # Extract only what we need: Date, Title, and Source
    news_data = []
    for art in articles['articles']:
        news_data.append({
            'date': art['publishedAt'],
            'title': art['title'],
            'source': art['source']['name'],
            'description': art['description']
        })

    return pd.DataFrame(news_data)
```

```python
# Checking without key using yFinance
def get_ticker_news(ticker_symbol):
    ticker = yf.Ticker(ticker_symbol)
    news_list = ticker.news

    # Check if news exists
    if not news_list:
        print(f"No news found for {ticker_symbol}")
        return pd.DataFrame()

    # Convert to DataFrame
    df = pd.DataFrame(news_list)

    # 2026 Fix: Print columns to see what Yahoo is currently sending
    print(f"Current Columns: {df.columns.tolist()}")

    # Select columns only if they exist, otherwise use 'title' as a fallback
    available_cols = [col for col in ['link', 'publisher', 'title'] if col in df.columns]

    if not available_cols:
        # If the expected columns are missing, just return the whole thing
        # so you can see what's inside
        return df

    return df[available_cols]
```

```python
# Ensure spaCy uses the GPU
is_using_gpu = spacy.prefer_gpu()
if is_using_gpu:
    print("GPU is active for spaCy!")
    # Use torch to confirm
    print(f"CUDA Device: {torch.cuda.get_device_name(0)}")

# Load the model
nlp = spacy.load("en_core_web_trf")
```

```python
def clean_financial_text(text):
    if not isinstance(text, str):
        return ""

    # Remove HTML tags (if scraped from web)
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove specific financial noise like "Ticker: " or "(Reuters)"
    text = re.sub(r'\([A-Z]+\.[A-Z]+\)', '', text) # Removes (AAPL.O)
    text = re.sub(r'Reuters|Brief|Headline', '', text, flags=re.IGNORECASE)
    # Standardize whitespace
    text = " ".join(text.split())

    return text

# Assuming df_apple is your dataframe from Phase 2
df_apple['clean_title'] = df_apple['title'].apply(clean_financial_text)
print("Text Cleaning Complete.")
```

```python
def extract_orgs(text):
    doc = nlp(text)

    # Extract entities labeled as ORG (Organizations)
    orgs = [ent.text for ent in doc.ents if ent.label_ == "ORG"]

    # Return unique organizations found in the string
    return list(set(orgs))

# Apply the NER model (This takes longer as it's a Transformer)
df_apple['companies_found'] = df_apple['clean_title'].apply(extract_orgs)

# Filter out rows where no company was found (optional)
df_filtered = df_apple[df_apple['companies_found'].map(len) > 0]
df_filtered.head()
```
